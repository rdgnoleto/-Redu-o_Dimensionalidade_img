```python
import urllib.request
import cv2
import numpy as np
import matplotlib.pyplot as plt

url = "https://thumbs.dreamstime.com/z/ilustra%C3%A7%C3%A3o-criativa-abstrata-com-le%C3%A3o-colorido-cor-preta-de-190846410.jpg?ct=jpeg"

# Ler imagem diretamente da URL (como array de bytes)
resp = urllib.request.urlopen(url)
img_array = np.asarray(bytearray(resp.read()), dtype=np.uint8)

# Decodificar imagem (formato BGR do OpenCV)
img_bgr = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# Converter para tons de cinza
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# Aplicar limiarização para converter em binária (preto e branco)
_, img_binary = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

# Exibir imagens
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.title("Original (BGR)")
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Tons de Cinza")
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Imagem Binária (P&B)")
plt.imshow(img_binary, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
